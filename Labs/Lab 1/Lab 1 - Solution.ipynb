### Linearly Spaced Points
There are many tasks that you need to generate a set of points in a range that are linearly spaced. For this task
we will use numpy modules arnage and linspace.

Note that the tutorial uses `np.arange`. While this is fine in some cases, we generally prefer to use `np.linspace`. `np.linspace(a, b, N)` divides the interval `[a, b]` into N points, while `np.arange(a, b, s)` will step from `a` to `b` with a fixed step size `s`.

One thing to keep in mind is that `np.linspace` always includes both end points while `np.arange` will *not* include the second end point `b`. For this reason, when we are plotting ranges of values we tend to prefer `np.linspace`.

Write two functions that return the same result.

```python
# create an array of linearly spaced points from -5(inclusive) to 6 (not inclusive) spaced by 1
array = np.arange(-5,6)
print(array)

# do the same with linspace
numlin_array = np.linspace(-5, 6, num=11, endpoint=False, dtype=float)
print(numlin_array)
```

Answer:

The function f(x,y) = x*y has a relative minimum at the bottoma as well. Also the critical points are x=0, y=0. However the shape is different and the end points(at the top) looks like they meet around the x and y values at -6 and 6 respectively.

```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return (x**2 + y**2 + 2*x + 2*y)

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

## Task 3 - Reading Documentation
We should also familiarize ourselves with looking up documentation and learning how to read it. Below is a section of code that plots a basic wireframe. Replace each `# Your answer here` with a description of what the line above does, what the arguments being passed in are, and how the arguments are used in the function. For example youe answer may look like,

`np.arange(2, 5, 0.2)`

`# This returns an array of numbers from 2 to 5 with an interval size of 0.2`

Hint: The `Shift` + `Tab` (to see what a function does) tip from earlier in the notebook may help here.

```python
from mpl_toolkits.mplot3d import axes3d

u = np.linspace(1.5*np.pi, -1.5*np.pi, 100)
#This creates a linearly spaced vectors where the start value is 1.5π 
#the end value is -1.5π, and the 100 is how many numbers will appear 
# between them
[x,y] = np.meshgrid(u, u)
# meshgrid will create a rectangular grid out of an array from the u values
# by assinging it to [x,y] we can see how it look like in a range
squared = np.sqrt(x.flatten()**2 + y.flatten()**2)
z = np.cos(squared)
# The squared variable will return a flattened one dimensional, flattening 
# the x any values and squaring them **2 = ^2, then z is getting the cosine
# value of the squared variable
z = z.reshape(x.shape)
# z is assigned the value of z.reshape which reshapes z into an array without chaning
# its value based on x.shape which will get the values of x and assign it to a tuple
# of array dimension, thus givings us our 3rd dimension

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# Now we delcare created our subplot using fig and making it 
# project in 3d, the subplot value is 111
ax.plot_wireframe(x, y, z, rstride=10, cstride=10)
# plot_wireframe is suppose to plot our values in 3d wireframe, however,
# it is missing the a 2 dimenional value and cannot proceed just yet
# x and y are their respetice 2d array data values, z is 3rd dimension and
# rstride is the array row stride and cstride is the array column stride
ax.view_init(elev=50., azim=30)
# This sets the elevation and azimuth of the axes, this can also be used to 
# rotate the axes, elev = stores the elevation angle, and azim store the 
# azimuth angle in the x,y plane
plt.savefig("figure1.png")
```
